```python
import time
from IPython.display import display

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Q_sigmoid(x, alpha=0.5, smooth=1):
    return np.around(1 / (1 + np.exp(-alpha * x)), decimals=smooth)
```

```python
x = np.linspace(-10, 20, 100)
y = Q_sigmoid(x)
plt.plot(x, y);
```

```python
y = x*Q_sigmoid(x, alpha=1)
plt.plot(x, y);
```

```python
fig = plt.figure(figsize=(16,8))
y = Q_sigmoid(x, alpha=0.5, smooth=2) + Q_sigmoid(x, alpha=0.5)
plt.plot(x, y);
```

```python
def Q_swish(x):
    return x * Q_sigmoid(x, alpha=1, smooth=1)

fig = plt.figure(figsize=(16,8))
y = Q_swish(x)
plt.plot(x, y);
```
